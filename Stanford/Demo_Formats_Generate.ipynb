```python
from pydub import AudioSegment
import time

def convert_to_flac(source_path, target_path):

    print(f"[{time.strftime('%Y-%m-%d %H:%M:%S')}] Loading audio file from: {source_path}")
    start_time = time.time()
    # Load the source file
    audio = AudioSegment.from_file(source_path)

    # Export as FLAC
    print(f"[{time.strftime('%Y-%m-%d %H:%M:%S')}] Converting to FLAC: {target_path}")
    audio.export(target_path, format="flac")

    end_time = time.time()
    print(f"[{time.strftime('%Y-%m-%d %H:%M:%S')}] Conversion completed. Duration: {end_time - start_time:.2f} seconds.")


# Path to your original audio file
source_path = "Stanford_NoQueen_Testing_10mins.wav"  # Change this to the path of your audio file
target_path = "Stanford_NoQueen_Testing_10mins.flac"  # Change this to your desired output path

# Convert to FLAC
convert_to_flac(source_path, target_path)

# Path to your original audio file
source_path = "Stanford_NoQueen_Training_30mins.wav"  # Change this to the path of your audio file
target_path = "Stanford_NoQueen_Training_30mins.flac"  # Change this to your desired output path

# Convert to FLAC
convert_to_flac(source_path, target_path)

# Path to your original audio file
source_path = "Stanford_Queen_Testing_10mins.wav"  # Change this to the path of your audio file
target_path = "Stanford_Queen_Testing_10mins.flac"  # Change this to your desired output path

# Convert to FLAC
convert_to_flac(source_path, target_path)

# Path to your original audio file
source_path = "Stanford_Queen_Training_30mins.wav"  # Change this to the path of your audio file
target_path = "Stanford_Queen_Training_30mins.flac"  # Change this to your desired output path

# Convert to FLAC
convert_to_flac(source_path, target_path)
```
