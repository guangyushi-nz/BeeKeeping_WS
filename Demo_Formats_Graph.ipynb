```python
import subprocess
import matplotlib.pyplot as plt
import numpy as np
import wave
from pydub import AudioSegment

# Function to convert audio file to WAV using ffmpeg
def convert_to_wav(audio_path, wav_path):
    AudioSegment.from_file(audio_path).export(wav_path, format="wav")

# Function to plot waveform
def plot_waveform(wav_path, title):
    with wave.open(wav_path, 'r') as wav_file:
        # Extract raw audio from WAV file
        signal = wav_file.readframes(-1)
        signal = np.frombuffer(signal, dtype="int16")
        # Get frame rate
        frate = wav_file.getframerate()
        # Calculate time for each sample
        time = np.linspace(0, len(signal) / frate, num=len(signal))
        # Plotting
        plt.figure(figsize=(12, 4))
        plt.plot(time, signal)
        plt.title(title)
        plt.ylabel('Amplitude')
        plt.xlabel('Time (s)')
        plt.show()

# Convert MP3 and FLAC to WAV
convert_to_wav('beeQueenSound_mp3_8K.mp3', 'beeQueenSound_mp3_8K.wav')
convert_to_wav('beeQueenSound.flac', 'beeQueenSound_flac.wav')

# Plot the waveforms
plot_waveform('beeQueenSound.wav', 'WAV Audio Waveform')
plot_waveform('beeQueenSound_mp3_8K.wav', 'MP3 Audio Waveform')
plot_waveform('beeQueenSound_flac.wav', 'FLAC Audio Waveform')
```

```python
import matplotlib.pyplot as plt
from scipy.io import wavfile
import numpy as np

def convert_to_mono(data):
    if len(data.shape) == 2:  # Check if the audio is stereo
        return data.sum(axis=1) / 2  # Convert to mono by averaging both channels
    return data

def plot_waveform(title, data, rate, start_sec, duration_sec):
    data_mono = convert_to_mono(data)
    start_sample = int(start_sec * rate)
    end_sample = int((start_sec + duration_sec) * rate)
    segment = data_mono[start_sample:end_sample]

    plt.figure(figsize=(10, 4))
    plt.plot(np.linspace(start_sec, start_sec + duration_sec, num=len(segment)), segment)
    plt.title(title)
    plt.ylabel('Amplitude')
    plt.xlabel('Time (s)')
    plt.tight_layout()
    plt.show()

# Read the WAV files
rate_wav, data_wav = wavfile.read('beeQueenSound.wav')
rate_mp3, data_mp3 = wavfile.read('beeQueenSound_mp3_8K.wav')
rate_flac, data_flac = wavfile.read('beeQueenSound_flac.wav')

# Plot waveform for a specific segment (e.g., 1 second starting at 10 seconds)
plot_waveform('WAV Waveform', data_wav, rate_wav, 10, 0.1)
plot_waveform('MP3 Waveform', data_mp3, rate_mp3, 10, 0.1)
plot_waveform('FLAC Waveform', data_flac, rate_flac, 10, 0.1)
```
