```python
from pydub import AudioSegment

def convert_to_flac(source_path, target_path):
    # Load the source file
    audio = AudioSegment.from_file(source_path)

    # Export as FLAC
    audio.export(target_path, format="flac")

# Path to your original audio file
source_path = "OSBH_Queen_Training_30mins.wav"  # Change this to the path of your audio file
target_path = "OSBH_Queen_Training_30mins.flac"  # Change this to your desired output path

# Convert to FLAC
convert_to_flac(source_path, target_path)

# Path to your original audio file
source_path = "OSBH_NoQueen_Training_30mins.wav"  # Change this to the path of your audio file
target_path = "OSBH_NoQueen_Training_30mins.flac"  # Change this to your desired output path

# Convert to FLAC
convert_to_flac(source_path, target_path)

# Path to your original audio file
source_path = "OSBH_Queen_Testing_10mins.wav"  # Change this to the path of your audio file
target_path = "OSBH_Queen_Testing_10mins.flac"  # Change this to your desired output path

# Convert to FLAC
convert_to_flac(source_path, target_path)

# Path to your original audio file
source_path = "OSBH_NoQueen_Testing_10mins.wav"  # Change this to the path of your audio file
target_path = "OSBH_NoQueen_Testing_10mins.flac"  # Change this to your desired output path

# Convert to FLAC
convert_to_flac(source_path, target_path)
```
