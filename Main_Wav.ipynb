```python
from scipy.signal import get_window
import IPython.display as ipd
import soundfile as sf
# create a mfcc and visualize 
import streamlit as st

import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt

def visualize_mel_spectrogram(audio_file, title):
    # Load audio file
    audio_data, sampling_rate = librosa.load(audio_file, sr=None)
    
    # Calculate Mel spectrogram
    mel_spectrogram = librosa.feature.melspectrogram(y=audio_data, sr=sampling_rate)
    log_mel_spectrogram = librosa.power_to_db(mel_spectrogram, ref=np.max)

    # Plot Mel spectrogram
    plt.figure(figsize=(6, 4))
    librosa.display.specshow(log_mel_spectrogram, sr=sampling_rate, x_axis='time', y_axis='mel', cmap='viridis')
    plt.colorbar(label='dB')
    plt.title(title)
    plt.tight_layout()
    plt.show()

# File paths
queen_wav_sample = "QueenBee_Testing_15mins.wav"
queen_mp3_64K_sample = "QueenBee_Testing_15mins_64K.mp3"
queen_mp3_16K_sample = "QueenBee_Testing_15mins_16K.mp3"
queen_flac_sample = "QueenBee_Testing_15mins.flac"

# Visualize Mel spectrogram for each audio file
visualize_mel_spectrogram(queen_wav_sample, 'Mel Spectrogram of QueenBee_Testing_15mins.wav')
visualize_mel_spectrogram(queen_mp3_64K_sample, 'Mel Spectrogram of QueenBee_Testing_15mins_64K.mp3')
visualize_mel_spectrogram(queen_mp3_16K_sample, 'Mel Spectrogram of QueenBee_Testing_15mins_16K.mp3')
visualize_mel_spectrogram(queen_flac_sample, 'Mel Spectrogram of QueenBee_Testing_15mins.flac')


```
